```python
# Dependencies
import pandas as pd 
```

```python
# load CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
coders_df=pd.read_csv(file, low_memory=False)
coders_df
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
extracted_df= coders_df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
extracted_df
```

```python
# Change reading 0 and 1 to No and Yes, respectively
extracted_df["AttendedBootcamp"] = extracted_df["AttendedBootcamp"].replace({0: "No", 1 : "Yes"})
extracted_df["AttendedBootcamp"].value_counts()
```

```python
# Extract rows for only those who attended a bootcamp
attended_bootcamp = extracted_df.loc[extracted_df['AttendedBootcamp'] == "Yes"]
attended_bootcamp.head()

bootcampattendees_df = attended_bootcamp.loc[attended_bootcamp["AttendedBootcamp"] == "Yes"]
bootcampattendees_df.head()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
#Create a DataFrame with two columns: One with the bootcamp name, and one with the number of respondents who went to each bootcamp.
Bootcampnames_df = bootcampattendees_df.groupby(["BootcampName"])
print(Bootcampnames_df)
NewBootcamp_df = Bootcampnames_df.count().sort_values("AttendedBootcamp", ascending=False).head()
NewBootcamp_df = NewBootcamp_df[["AttendedBootcamp"]].head()
NewBootcamp_df = NewBootcamp_df.rename(columns= {"AttendedBootcamp": "Number of Attendees" })
NewBootcamp_df
```

```python
bootcampattendees_df["BootcampRecommend"] = bootcampattendees_df["BootcampRecommend"].replace({0: "No", 1 : "Yes"})
bootcampattendees_df["BootcampRecommend"].value_counts()
```

```python
significant_bootcamp_df = bootcampattendees_df[bootcampattendees_df["BootcampRecommend"] == "Yes"]
significant_bootcamp_df.head()
```

```python
significant_bootcamp_df["BootcampRecommend"] = significant_bootcamp_df["BootcampRecommend"].replace({0: "No", 1 : "Yes"})
significant_bootcamp_df["BootcampRecommend"].value_counts()

recommend_bootcamp = significant_bootcamp_df[significant_bootcamp_df["BootcampRecommend"] == "Yes"]
recommend_bootcamp.head()
```

```python
#* Create a DataFrame with two columns: One with the bootcamp name, and one with the number of respondents who went to each bootcamp.
bootcamp_recommend_df = recommend_bootcamp.groupby(["BootcampName"])
print(bootcamp_recommend_df)
ReccBootcamp_df = bootcamp_recommend_df.count().sort_values("BootcampRecommend", ascending=False).head()
ReccBootcamp_df = ReccBootcamp_df[["BootcampRecommend"]].head()
ReccBootcamp_df = ReccBootcamp_df.rename(columns= {"BootcampRecommend": "Number of Attendees who Recommend this Bootcamp" })
ReccBootcamp_df
```

```python
# Merge the two created data frames on the name of tbe bootcamp
# Calculate percentage of each bootcamp's students who are recommenders
merge_df = pd.merge(NewBootcamp_df, ReccBootcamp_df, on="BootcampName").reset_index()
merge_df
merge_df["Percentage"] = (merge_df["Number of Attendees who Recommend this Bootcamp"] / merge_df["Number of Attendees"]) * 100
merge_df
merge_df = merge_df.style.format({"Percentage": "{:,.1f}%"})
merge_df
```

```python
# Export to excel and remove index
merge_df.to_excel("Output/PandasPart2.xlsx", index=False)
```
